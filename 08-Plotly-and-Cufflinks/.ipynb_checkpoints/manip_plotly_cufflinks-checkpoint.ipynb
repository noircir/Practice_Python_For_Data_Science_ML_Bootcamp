```python
# DATA
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({ 'Category': ['A', 'B', 'C'], 'Values': [32,48,50]})
```

```python
df2
```

```python
df.plot()
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers')
```

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
df.head()
```

```python
# Not useful: a bar for each data point
df.iplot(kind='bar')
```

```python
# Aggregate
# Not very useful, neither: 100 of each of them
df.count().iplot(kind='bar')
```

```python
# iplot is good for some kind of aggregation
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
# 3D surface plot

df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[500,400,300,200,100]})
```

```python
df3.head()
```

```python
df3.iplot(kind='surface')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
```

```python
df3.iplot(kind='surface')
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
df3.iplot(kind='surface',colorscale='YlGnBu')
```

```python
df['A'].iplot(kind='hist',bins=50)
```

```python
df.iplot(kind='hist')
```

```python
# Spread (for stock data)
df[['A','B']].iplot(kind='spread')
```

```python
# Bubble plot (world GDP, happiness, ets)
df.iplot(kind='bubble',x='A',y='B',size='C')
```

```python
# scatter matrix (like Seaborn's pairplot)
df.scatter_matrix()
```
