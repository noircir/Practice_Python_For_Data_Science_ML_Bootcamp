```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# centers=4 will create 4 clusters, and 4 categories of labels
data = make_blobs(n_samples=400,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
# data is tuple, the first element is ndarray
data[0].shape
```

```python
# Labels for clustering
data[1].shape
```

```python
# scatterplot for 2 features
plt.scatter(data[0][:,0], data[0][:,1])
```

```python
# the same scatterplot but with 4 centers 

plt.scatter(data[0][:,0], data[0][:,1],c=data[1])
```

```python
plt.scatter(data[0][:,0], data[0][:,1],c=data[1],cmap='rainbow')
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
# Predicted labels. IRL we wouldn't be able to compare, but since we created the blob labels,
# we have something to compare to.
kmeans.labels_
```

```python
# 1 row, 2 columns, shared Y axis, 
sns.set_style('whitegrid')
fig, (subfigure1,subfigure2) = plt.subplots(1,2, sharey=True,figsize=(10,6))

subfigure1.set_title('K Means')
# scatterplot of data's 2 features; centers (labels) from our model
subfigure1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_,cmap='viridis')

subfigure2.set_title('Original')
# scatterplot of the same 2 features; centers (labels) from blob labels
subfigure2.scatter(data[0][:,0],data[0][:,1], c=data[1],cmap='rainbow')
```
